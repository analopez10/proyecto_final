```python
import pandas as pd
from itertools import combinations
from collections import Counter

# Suponiendo que tu DataFrame de vinos se llama wines

# Eliminar corchetes y comillas de la columna 'Grapes'
wines['Grapes'] = wines['Grapes'].str.replace("[\[\]']", '', regex=True)

# Crear una lista de listas de uvas para cada vino
wines['Grapes_List'] = wines['Grapes'].str.split(', ')

# Contar todas las combinaciones posibles de uvas
all_combinations = []
for grapes in wines['Grapes_List']:
    if len(grapes) > 1:
        all_combinations.extend(combinations(sorted(grapes), 2))

# Contar la frecuencia de cada combinaci√≥n de uvas
combination_counts = Counter(all_combinations)

# Crear un DataFrame con las combinaciones y sus frecuencias
combination_df = pd.DataFrame(combination_counts.items(), columns=['Grape Combination', 'Count'])
combination_df['Grape 1'], combination_df['Grape 2'] = zip(*combination_df['Grape Combination'])
combination_df.drop(columns=['Grape Combination'], inplace=True)

# Mostrar las primeras filas del DataFrame resultante
print(combination_df.head())
```

```python
# Crear una tabla cruzada
cross_table = combination_df.pivot_table(index='Grape 1', columns='Grape 2', values='Count', fill_value=0)

# Filtrar las combinaciones que son todas ceros
cross_table = cross_table.loc[:, (cross_table != 0).any(axis=0)]

# Mostrar las primeras filas de la tabla cruzada
cross_table.head()
```

```python
import pandas as pd

# Combinar los dataframes en base a 'RegionID'
combined_df = pd.merge(clima, wines, on='RegionID')

# Verificar los datos combinados
combined_df.head()
```

```python
# Seleccionar las columnas relevantes
climate_vars = ['tmin', 'tmax', 'prcp', 'snow', 'pres', 'tsun']
wine_chars = ['ABV', 'Body', 'Acidity']

# Crear un nuevo dataframe desglosado
wines_climate_list = []

for climate_var in climate_vars:
    for wine_char in wine_chars:
        temp_df = combined_df[['RegionID', climate_var, wine_char]].copy()
        temp_df.columns = ['RegionID', 'ClimateVariable', 'WineCharacteristic']
        temp_df['ClimateVariableName'] = climate_var
        temp_df['WineCharacteristicName'] = wine_char
        wines_climate_list.append(temp_df)

wines_climate = pd.concat(wines_climate_list, axis=0, ignore_index=True)

wines_climate.head()
```

```python
# Crear una lista de a√±os de cosecha √∫nicos en el dataset de vinos
year_of_ref = 2023

# Filtrar el dataframe de clima para incluir solo esos a√±os
filtered_clima = clima[clima['time']== year_of_ref ]
filtered_clima.head()
```

```python
# Combinar los DataFrames en base a 'RegionID'
combined_df_filtered = pd.merge(filtered_clima, wines, left_on='RegionID', right_on='RegionID')
```

#### ANOVA para Body y Acidity

```python
import pandas as pd
import scipy.stats as stats

# ANOVA para Body
for climate_var in climate_vars:
    f_val, p_val = stats.f_oneway(
        combined_df_filtered[combined_df_filtered['Body'] == 'Light-bodied'][climate_var],
        combined_df_filtered[combined_df_filtered['Body'] == 'Medium-bodied'][climate_var],
        combined_df_filtered[combined_df_filtered['Body'] == 'Full-bodied'][climate_var]
    )
    print(f'ANOVA results for {climate_var} and Body: F-value = {f_val}, p-value = {p_val}')

# ANOVA para Acidity
for climate_var in climate_vars:
    f_val, p_val = stats.f_oneway(
        combined_df_filtered[combined_df_filtered['Acidity'] == 'Low'][climate_var],
        combined_df_filtered[combined_df_filtered['Acidity'] == 'Medium'][climate_var],
        combined_df_filtered[combined_df_filtered['Acidity'] == 'High'][climate_var]
    )
    print(f'ANOVA results for {climate_var} and Acidity: F-value = {f_val}, p-value = {p_val}')
```

#### Chi-cuadrado for Body and Acidity

```python
# Tablas de contingencia para Chi-cuadrado
contingency_table_body = pd.crosstab(combined_df_filtered['Body'], combined_df_filtered['Acidity'])

# Prueba de Chi-cuadrado
chi2, p, dof, ex = stats.chi2_contingency(contingency_table_body)
print(f'Chi-cuadrado test results for Body and Acidity: Chi2 = {chi2}, p-value = {p}')
```

Significado de los Resultados

	1.	Chi2 Statistic (Valor Chi-cuadrado)
	‚Ä¢	Un valor Chi2 alto indica una fuerte evidencia de asociaci√≥n entre las variables.
	‚Ä¢	En estos resultados, los valores de Chi2 son muy altos, lo que sugiere una fuerte relaci√≥n entre las variables comparadas.
	2.	p-value (Valor p)
	‚Ä¢	El valor p mide la probabilidad de que cualquier asociaci√≥n observada entre las variables haya ocurrido por azar.
	‚Ä¢	Un valor p de 0.0 indica que la probabilidad de que la asociaci√≥n observada sea debida al azar es extremadamente baja.
	‚Ä¢	Normalmente, un valor p inferior a 0.05 se considera estad√≠sticamente significativo. En estos resultados, todos los valores p son 0.0, lo que indica una fuerte significancia estad√≠stica.



Conclusi√≥n

	1.	ABV y Body
	‚Ä¢	Existe una fuerte asociaci√≥n entre el contenido de alcohol (ABV) y el cuerpo del vino. Esto sugiere que el contenido de alcohol puede influir en la percepci√≥n del cuerpo del vino o viceversa.
	2.	ABV y Acidity
	‚Ä¢	Tambi√©n hay una fuerte asociaci√≥n entre el contenido de alcohol (ABV) y la acidez del vino. Esto podr√≠a implicar que el nivel de alcohol afecta la acidez percibida del vino o que ciertos niveles de acidez son m√°s comunes en vinos con ciertos niveles de alcohol.
	3.	Body y Acidity
	‚Ä¢	Existe una fuerte asociaci√≥n entre el cuerpo del vino y su acidez. Esto sugiere que los vinos con diferentes niveles de cuerpo tienden a tener diferentes niveles de acidez.

	‚Ä¢	Dado que hay una fuerte correlaci√≥n entre estas variables, es posible que se necesiten t√©cnicas espec√≠ficas para manejar la multicolinealidad en modelos predictivos.


#### ANOVA para cada variable clim√°tica con ABV

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Filtrar el dataframe de clima para incluir solo el a√±o de referencia
year_of_ref = 2023
filtered_clima = clima[clima['time'] == year_of_ref]

# Asegurarse de que ABV est√© en el DataFrame y no contenga valores nulos
wines_climate = wines.merge(filtered_clima, on='RegionID', how='inner').dropna(subset=['ABV'])

# Variables clim√°ticas
climatic_vars = ['tmin', 'tmax', 'prcp', 'snow', 'pres', 'tsun']

# Convertir variables clim√°ticas a categ√≥ricas si no lo est√°n
for var in climatic_vars:
    wines_climate[var] = pd.cut(wines_climate[var], bins=10, labels=False)

# ANOVA para cada variable clim√°tica con ABV
anova_results = {}

for var in climatic_vars:
    formula = f'ABV ~ C({var})'
    model = ols(formula, data=wines_climate).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    anova_results[var] = anova_table

# Crear diccionario con los resultados del ANOVA
anova_results_dict = {
    'Variable': [],
    'Sum of Squares': [],
    'Degrees of Freedom': [],
    'F-value': [],
    'P-value': []
}

for var, result in anova_results.items():
    anova_results_dict['Variable'].append(var)
    anova_results_dict['Sum of Squares'].append(result['sum_sq'][0])
    anova_results_dict['Degrees of Freedom'].append(result['df'][0])
    anova_results_dict['F-value'].append(result['F'][0])
    anova_results_dict['P-value'].append(result['PR(>F)'][0])

# Crear DataFrame con los resultados del ANOVA
anova_results_df = pd.DataFrame(anova_results_dict)

# Mostrar la tabla de resultados del ANOVA
anova_results_df
```
